```python
import random
win_rate = {0: 7, 1: 6, 2: 5, 3: 4, 4: 3, 5: 2, 6: 1}
cycle = 1
while True:
    lottos = []
    win_num = []
    lego = input("자동, 수동 혹은 복합 중 원하시는 복권 구매 방식을 선택해주세요.: ")
    if lego == "복합":
        auto, man = map(int, input("자동으로 구매할 복권의 매수와 수동으로 구매할 복권의 매수를 입력해주세요. (최대 합은 5입니다.): ").split())
        for i in range(auto):
            numbers = []
            while len(numbers) < 6:
                num = random.randint(1, 45)
                if num in numbers:
                    pass
                else:
                    numbers.append(num)
            lottos.append(numbers)
        for i in range(man):
            print(f"{i+auto+1}번째 복권")
            numbers = []
            while len(numbers) < 6:
                num1, num2, num3, num4, num5, num6 = map(int, input("1부터 45 사이의 수를 6개 입력하세요. (중복 불가): ").split())
                numbers = [num1, num2, num3, num4, num5, num6]
            lottos.append(numbers)
        print("\n[구매하신 복권]\n", )
        for i in lottos:
            print(i)

    elif lego == "자동":
        auto = int(input("구매하실 자동복권의 매수를 입력해주세요. (최대 5매): "))
        for i in range(auto):
            numbers = []
            while len(numbers) < 6:
                num = random.randint(1, 45)
                if num in numbers:
                    pass
                else:
                    numbers.append(num)
            lottos.append(numbers)
        print("\n[구매하신 복권]\n", )
        for i in lottos:
            print(i)
    else:
        man = int(input("구매하실 수동복권의 매수를 입력해주세요. (최대 5매): "))
        for i in range(man):
            print(f"{i+1}번째 복권")
            numbers = []
            while len(numbers) < 6:
                num1, num2, num3, num4, num5, num6 = map(int, input("1부터 45 사이의 수를 6개 입력하세요. (중복 불가): ").split())
                numbers = [num1, num2, num3, num4, num5, num6]
            lottos.append(numbers)
        print("\n[구매하신 복권]\n", )
        for i in lottos:
            print(i)

    while len(win_num) < 6:
        num = random.randint(1, 45)
        if num in win_num:
            pass
        else:
            win_num.append(num)
    print("\n[당첨 번호]\n", win_num)

    ratings = []
    for i in lottos:
        count = 0
        for j in range(len(i)):
            if i[j] in win_num:
                count += 1
        ratings.append(count)
    max_rating = max(ratings)

    print("\n", max_rating, "개 맞은 복권이 있습니다.!")
    if max_rating == 0:
        print("아쉽네요 ㅠㅠ 꽝입니다. ㅠㅠ")
    else:
        print(f"{win_rate[max_rating]}등입니다! 축하합니다!")
        print("\n")
    
    break
```
